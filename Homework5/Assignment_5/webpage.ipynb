# Facebook SNAP Dataset 

```python
import tarfile
import pandas as pd
from collections import Counter

# Path to your tar.gz (adjust if needed)
tar_path = "facebook.tar.gz"

with tarfile.open(tar_path, "r:gz") as tar:
    # Pick one ego network; here we use ego 0
    edges_member = tar.getmember("facebook/0.edges")
    f = tar.extractfile(edges_member)

    # Read the edge list (space-separated integers)
    pairs = []
    for line in f:
        line = line.decode().strip()
        if not line:
            continue
        u, v = map(int, line.split())
        pairs.append((u, v))

# Compute degrees
deg = Counter()
for u, v in pairs:
    deg[u] += 1
    deg[v] += 1

df_nodes = pd.DataFrame({
    "node_id": list(deg.keys()),
    "degree": list(deg.values())
})

def degree_category(d):
    if d <= 10:
        return "Low (≤10)"
    elif d <= 30:
        return "Medium (11–30)"
    else:
        return "High (>30)"

df_nodes["degree_category"] = df_nodes["degree"].apply(degree_category)

# Save this tidy node-level dataset to CSV;
# commit this CSV to your GitHub repo (e.g. in data/facebook_ego0_nodes.csv)
df_nodes.to_csv("facebook_ego0_nodes.csv", index=False)
```
